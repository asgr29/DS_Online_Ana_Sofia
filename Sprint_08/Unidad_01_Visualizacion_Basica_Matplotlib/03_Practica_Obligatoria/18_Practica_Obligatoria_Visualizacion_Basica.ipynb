 1-Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia) 

 

```python
#Datos
top_ciudades = ["Los Angeles", "San Diego", "Sacramento ciudades", "Anaheim", "Santa Ana"]
poblacion =  [3884.307, 1345.895, 466.488, 336.0, 324.528]
# Creacin grfico de barras
fig,ax = plt.subplots(figsize= (10,8))
ax.bar(top_ciudades, poblacion, color=["blue", "gray", "olive", "brown"], alpha=0.5)
for i in range(len(top_ciudades)):
    ax.text(i, poblacion[i] + 50, f"{poblacion[i]:.3f} mil", ha="center", fontsize=12)
ax.set_title("Ciudades ms pobladas de California", fontsize=12)
ax.set_ylabel("Habitantes (miles)")
ax.set_xlabel("Ciudades");
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")

```python
fig,ax = plt.subplots(figsize= (10,8))
ax.hist(df["area_total_km2"], bins= 50, color= "steelblue", label= "Hist area")
ax.set_xlabel("Nmero de ciudades", fontname="Bradley Hand ITC", fontsize=18,)
ax.set_ylabel("Frecuencia (Km2)", fontname="Bradley Hand ITC", fontsize=18)
ax.legend();
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python

x = df["area_water_km2"]
y = df["population_total"] / 1000
ciudades = df["city"]
fig,ax = plt.subplots(figsize= (10,8))
ax.scatter(x, y, c= "green", alpha= 0.5)
ax.set_xlabel("rea de agua en la ciudad(km2)")
ax.set_ylabel("Poblacin(miles habitantes)")
ax.set_title("Poblacin por kilometro cuadrado de agua")
# Coordenadas de los outliers extremos
max_x = x.max()
max_y = y.max()
# Coordenada donde estar el texto "Outliers"
texto_x = 45
texto_y = 3500
#Flechas desde el texto hacia los puntos extremos
ax.annotate("Outliers",
             xy=(texto_x, texto_y),  #posicin del texto
             xytext=(texto_x, texto_y),  #mismo punto para que el texto est ah
             fontsize=14, color="black")
#Flecha hacia el punto con mayor rea de agua
ax.annotate("",
             xy=(max_x, y[x.idxmax()]),
             xytext=(texto_x, texto_y),
             arrowprops=dict(arrowstyle="->", color="black"))
#Flecha hacia el punto con mayor poblacin
ax.annotate("",
             xy=(x[y.idxmax()], max_y),
             xytext=(texto_x, texto_y),
             arrowprops=dict(arrowstyle="->", color="black"));
```

4. El boxplot de la altura de las ciudades.

```python
fig,ax = plt.subplots(figsize= (10,8))
ax.boxplot(df["elevation_m"], vert= False, patch_artist=True,
boxprops=dict(facecolor="red", color="black"), medianprops=dict(color="black")) #Facecolor es el relleno de la caja
ax.set_xlabel("Altitud(m)")
ax.set_title("Diagrama de caja de la Altitud de las ciudades de California");
```
